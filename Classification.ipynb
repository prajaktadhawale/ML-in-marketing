```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df_training = pd.read_csv("https://raw.githubusercontent.com/miltondavila/datasets/main/iris_training.csv")
df_training
```

```python
# Make sure data types are correct (no objects)
df_training.dtypes # convert in assignment
```

```python
# Pop label
target = df_training.pop('species')
```

```python
# Define features

numeric_feature_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
numeric_features = df_training[numeric_feature_names]
numeric_features.head()
```

```python
# Convert data to Tensorflow tensor

numeric_features = tf.convert_to_tensor(numeric_features)
numeric_features
```

```python
# Normalize data

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(numeric_features)
```

```python
# Define the model

def get_basic_model():
  model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3) # Possible labels 0, 1 or 2
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                # loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
# Train model

BATCH_SIZE = 128

model = get_basic_model()
model.fit(numeric_features, target, epochs=500, batch_size=BATCH_SIZE)
#output is affected according to number of epochs, experiment with it
#accuracy should be maximum with loss minimum
#balance the two
```

```python

# Download test data

df_test = pd.read_csv("https://raw.githubusercontent.com/miltondavila/datasets/main/iris_test.csv")
df_test.head()
```

```python
# Pop label
test_target = df_test.pop('species')
```

```python
# Define test features

test_numeric_feature_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
test_numeric_features = df_test[test_numeric_feature_names]
test_numeric_features.head()
```

```python
# Convert test data to Tensorflow tensor

test_numeric_features = tf.convert_to_tensor(test_numeric_features)
test_numeric_features
```

```python
# Normalize data

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(test_numeric_features)
```

```python
# Evaluate test data based on training data

score = model.evaluate(test_numeric_features, test_target, verbose=1)
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')
```

```python

# Save model (optional)

model.save('iris_model')
```

```python
# Download cases to be predicted

df_predict = pd.read_csv("https://raw.githubusercontent.com/miltondavila/datasets/main/iris_predict.csv")
df_predict
```

```python

# Make sure data types are correct (no objects)
# convert it into floats
df_predict.dtypes
```

```python
# Convert data to Tensorflow tensor

predict_numeric_features = tf.convert_to_tensor(df_predict)
predict_numeric_features
```

```python
# Normalize

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(predict_numeric_features)
```

```python
# Predict labels

class_names = ['Setosa', 'Virginica', 'Versicolor']

predictions = model(predict_numeric_features, training=False)

# Create new columns in dataframe
df_predict['label'] = None
df_predict['certainty'] = None

for i, logits in enumerate(predictions):
  class_idx = tf.argmax(logits).numpy()
  p = tf.nn.softmax(logits)[class_idx]
  name = class_names[class_idx]
  print("Example {} prediction: {} ({:4.1f}%)".format(i, name, 100*p))

  # Save predictions to dataframe
  df_predict["label"].iloc[i] = name
  df_predict['certainty'].iloc[i] = format(p)
```
